````python
# 3. ESSENTIAL EDA WITH FIGURE SAVING
print("Running Essential EDA...\n")
logging.info("Running essential EDA and saving figures.")

# EDA figures folder path
eda_fig_folder = os.path.join(PROJECT_ROOT, "reports/eda_figures")
os.makedirs(eda_fig_folder, exist_ok=True)

df= telco_df.copy()
# Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "churn_distribution.png"))
plt.show()
plt.close()
logging.info("Churn distribution plot saved.")

print("Churn Percentage (%):")
print(df['Churn'].value_counts(normalize=True).mul(100).round(2), "\n")

# Contract Type vs Churn
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title('Contract Type vs Churn')
plt.xlabel('Contract')
plt.ylabel('Customer Count')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "contract_vs_churn.png"))
plt.show()
plt.close()
logging.info("Contract vs Churn plot saved.")

# Hidden Patterns / Grouped Analysis
grouped = telco_df.groupby(['Contract','Churn'])['customerID'].count().reset_index()
grouped.rename(columns={'customerID':'Count'}, inplace=True)
print("\n Hidden Patterns: Contract vs Churn")
display(grouped)

# Internet Service vs Churn
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set1')
plt.title('Internet Service vs Churn')
plt.xlabel('InternetService')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "internet_service_vs_churn.png"))
plt.show()
plt.close()
logging.info("Internet Service vs Churn plot saved.")

# Tenure Distribution by Churn (Customer Loyalty Insight, KDE Plot)
plt.figure(figsize=(6,4))
sns.kdeplot(data=df, x='tenure', hue='Churn', fill=True, common_norm=False)
plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (months)')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "tenure_distribution_by_churn.png"))
plt.show()
plt.close()
logging.info("Tenure distribution plot saved.")

# Monthly Charges vs Churn (Billing Pressure Insight, Boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Set3')
plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "monthly_charges_vs_churn.png"))
plt.show()
plt.close()
logging.info("Monthly charges boxplot saved.")

# Essential Correlation Heatmap (numeric + target only)
df_corr = df[['SeniorCitizen', 'tenure', 'MonthlyCharges']].copy()
df_corr['Churn'] = df['Churn'].map({'No':0, 'Yes':1})

plt.figure(figsize=(6,4))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Essential)')
plt.tight_layout()
plt.savefig(os.path.join(eda_fig_folder, "correlation_heatmap.png"))
plt.show()
plt.close()
logging.info("Correlation heatmap saved.")

# Outliers check with boxplots
numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

for col in numeric_cols:
    print(f"\nChecking outliers for: {col}")

    # Try converting to numeric (for plotting only)
    try:
        numeric_series = pd.to_numeric(telco_df[col], errors="coerce")

        # Boxplot
        plt.figure(figsize=(6,4))
        sns.boxplot(x=numeric_series)
        plt.title(f"Boxplot of {col}")
        plt.xlabel(col)
        plt.grid(True)
        plt.tight_layout()

        save_path = os.path.join(eda_fig_folder, f"{col}_boxplot.png")
        plt.savefig(save_path)
        plt.show()
        plt.close()

        logging.info(f"{col} boxplot saved to {save_path}")

        # Outlier count 
        Q1 = numeric_series.quantile(0.25)
        Q3 = numeric_series.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = numeric_series[(numeric_series < lower) | (numeric_series > upper)]
        outcount = outliers.notna().sum()

        print(f"Outliers Found: {outcount}")

    except Exception as e:
        print(f" Could not process {col}: {e}")
        logging.error(f"Error in boxplot outlier check for {col}: {e}")

print("\n Boxplot outlier check completed.")
logging.info(" Boxplot outlier check completed.")

# FINAL STATUS
print("\n EDA Completed Successfully! All essential plots saved.")
logging.info("Essential EDA fully completed and all figures saved.")
````

````python
# 7B. Model VISUALIZATION  

# Create Visuals Directory
VISUALS_DIR = os.path.join(OUTPUTS_DIR, "models_visuals")
os.makedirs(VISUALS_DIR, exist_ok=True)
print(f"Visuals will be saved in: {VISUALS_DIR}")

# Convert X_test_ml ‚Üí numpy for plotting
X_test_ml_arr = X_test_ml.toarray() if hasattr(X_test_ml, "toarray") else np.array(X_test_ml)

# ROC-AUC CURVE (ML ONLY)
def plot_roc_auc(models_dict, X_test, y_test, save_path):
    plt.figure(figsize=(8, 6))

    for name, model in models_dict.items():
        if model is None:
            continue
        try:
            y_prob = model.predict_proba(X_test)[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")
        except:
            continue  # skip if model has no probability output

    plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
    plt.title("ROC-AUC Curve Comparison (ML Models)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.tight_layout()

    plt.savefig(save_path)
    plt.show()
    plt.close()

# CORRECT MODEL ACCESS (FROM ml_results)
plot_roc_auc(
    models_dict={
        "Logistic Regression": ml_results["ml_results"].get("Logistic Regression", {}).get("model", None),
        "Random Forest": ml_results["ml_results"].get("Random Forest", {}).get("model", None),
        "XGBoost": ml_results["ml_results"].get("XGBoost", {}).get("model", None),
    },
    X_test=X_test_ml_arr,
    y_test=y_test_ml,
    save_path=os.path.join(VISUALS_DIR, "roc_auc_comparison.png")
)
print(" ROC-AUC Curve Saved!")

# CONFUSION MATRIX (BEST ML)
best_ml_model = ml_results["best_ml_model"]
y_pred = best_ml_model.predict(X_test_ml_arr)
cm = confusion_matrix(y_test_ml, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(f"Confusion Matrix ‚Äì {ml_results['best_ml_name']}")
plt.savefig(os.path.join(VISUALS_DIR, "confusion_matrix_best_ml.png"))
plt.show()
plt.close()
print(" Confusion Matrix Saved!")

# DL LOSS CURVE (IF HISTORY EXISTS)
# Load DL training log (CSVLogger output)
csv_log_path = os.path.join(OUTPUTS_DIR, "dl_training_log.csv")
try:
    history_log = pd.read_csv(csv_log_path)
    print("dl_training_log.csv loaded successfully from:", csv_log_path)
    print(history_log.head())  # preview first few rows
except FileNotFoundError:
    print(" dl_training_log.csv not found. Train DL model with CSVLogger first.")
    history_log = None

# Plot Loss Curve
if history_log is not None:
    plt.figure(figsize=(7,5))
    plt.plot(history_log["loss"], label="Train Loss")
    plt.plot(history_log["val_loss"], label="Val Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("DL Training Loss Curve")
    plt.legend()
    
    loss_path = os.path.join(VISUALS_DIR, "dl_loss_curve.png")
    plt.savefig(loss_path)
    plt.show()
    plt.close()
    print(" Loss Curve Saved!")

    # Plot Accuracy (only if available)
    if "accuracy" in history_log.columns and "val_accuracy" in history_log.columns:
        plt.figure(figsize=(7,5))
        plt.plot(history_log["accuracy"], label="Train Accuracy")
        plt.plot(history_log["val_accuracy"], label="Val Accuracy")
        plt.xlabel("Epochs")
        plt.ylabel("Accuracy")
        plt.title("DL Training Accuracy Curve")
        plt.legend()

        acc_path = os.path.join(VISUALS_DIR, "dl_accuracy_curve.png")
        plt.savefig(acc_path)
        plt.show()
        plt.close()
        print(f" Accuracy Curve Saved!")
    else:
        print(" Accuracy columns not found ‚Äî skipping accuracy plot.")

# XGBoost FEATURE IMPORTANCE (BEST ML MODEL)
try:
    if ml_results["best_ml_name"] == "XGBoost":
        model = ml_results["best_ml_model"]

        # Get feature names
        if hasattr(X_train_ml, 'columns'):
            feature_names = X_train_ml.columns.tolist()
        else:
            feature_names = [f"f{i}" for i in range(X_train_ml.shape[1])]

        # Extract importance values
        importance = model.feature_importances_
        fi = pd.DataFrame({"Feature": feature_names, "Importance": importance})
        fi = fi.sort_values(by="Importance", ascending=False).head(20)  # Top 20

        plt.figure(figsize=(8, 6))
        plt.barh(fi["Feature"], fi["Importance"])
        plt.gca().invert_yaxis()  # Highest importance on top
        plt.xlabel("Importance")
        plt.title("Top 20 XGBoost Feature Importances")
        save_path = os.path.join(VISUALS_DIR, "feature_importance_xgboost.png")
        plt.savefig(save_path)
        plt.show()
        plt.close()
        print(f" Feature Importance saved at: {save_path}")
    else:
        print(f" Best model is {ml_results['best_ml_name']} ‚Äî Skipping XGBoost plot")

except Exception as e:
    print(" Feature Importance Error:", e)

print("\n ALL VISUALIZATIONS COMPLETE!")
````
